```python
import pandas as pd
import numpy as np
results = pd.read_csv("./data/school_res.csv", sep=',', encoding='cp1251')
results['М_ИТОГ'] = results['М_ИТОГ'].astype(int)
```

#### Итог и сдал

```python
results["ИТОГ"] = results['М_ИТОГ']+results['А_ИТОГ']+results['Р_ИТОГ']
results['СДАЛ'] = (results['М_ИТОГ'] >= 8) & (results['А_ИТОГ'] >= 4) & (results['Р_ИТОГ'] >= 4)
display(results.head())
```

#### Cортировка по итоам

```python
display(results.sort_values(["ИТОГ"], ascending=[0]).iloc[0:20, :])
```

#### Сдавшие

```python
good = results[(results["М_ИТОГ"] >= 8) & (results['А_ИТОГ'] >= 4) & (results['Р_ИТОГ'] >= 4)].sort_values(['ИТОГ'], ascending=[0])

print(f"Всего из {len(results)} учеников прошло: {len(good)}\nПроцент сдавших: {round(len(good)/len(results)*100, 2)}% ({len(good)}/{len(results)})")
display(good[['Шифр', 'М_ИТОГ', 'А_ИТОГ', 'Р_ИТОГ', 'ИТОГ']])
```

#### Статистика

```python
math_average = round(sum(results['М_ИТОГ'])/len(results['М_ИТОГ']), 1)
eng_average = round(sum(results['А_ИТОГ'])/len(results['А_ИТОГ']), 1)
rus_average = round(sum(results['Р_ИТОГ'])/len(results['Р_ИТОГ']), 1)

print(f"Средний балл по предметам:\nматематика: {math_average}\nанглийский: {eng_average}\nрусский: {rus_average}")
```
