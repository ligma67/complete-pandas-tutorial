Setting columns with a condition

```python
import numpy as np
#eeffoc['new_price'] = np.where(eeffoc["Coffee Type"] == "Espresso", 4.99, 2.99)
eeffoc['new_price'] = np.where(eeffoc['Day'].isin(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]), 3.99, 5.99)
```

Removing, renaming and combining columns

```python
eeffoc.drop(columns="price", inplace=True)

eeffoc["revenue"] = eeffoc["Units Sold"] * eeffoc["new_price"]

eeffoc = eeffoc.rename(columns={'new_price':"price"})
eeffoc
```

String operations with data & DT

```python
bios_copy = bios.copy()
bios_copy["first_name"] = bios_copy["name"].str.split(' ').str[0]
bios_copy["born_year"] = bios_copy["born_date"].str[:4]
bios_copy["born_datetime"] = pd.to_datetime(bios_copy["born_date"], format="%Y-%m-%d") #errors='coerce'
bios_copy["born_year"] = bios_copy["born_datetime"].dt.year
bios_copy["born_month"] = bios_copy["born_datetime"].dt.month_name()
display(bios_copy[["athlete_id", "first_name", "born_year", "born_datetime"]])
bios_copy[bios_copy["born_datetime"].dt.is_leap_year]
```

Saving DataFrame

Add/Remove with LAMBDA & custom FUNCTIONS / ft .apply()

```python
bios["height_category"] = bios["height_cm"].apply(lambda x: "High" if x > 190 else ("Medium" if x > 170 else "Short"))
bios.head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175:
        return "Short"
    elif row["height_cm"] < 190:
        return "Medium"
    else:
        return "High"
bios["height_cat"] = bios.apply(categorize_athlete, axis=1)  #you should always specify tf youre doing

bios.head()
```
