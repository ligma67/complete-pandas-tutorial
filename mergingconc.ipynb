# **Merging & Concatenating DATA**

```python
import pandas as pd
bios = pd.read_csv("./data/bios.csv")
nocs = pd.read_csv("./data/noc_regions.csv")
results = pd.read_csv("./data/results.csv")
```

### Мини проект по медианным зарплатам

```python
salaries = pd.read_csv("./data/median-wages.csv", encoding='cp1251', sep=';')
salaries["date"] = pd.to_datetime(salaries["date"], format="%Y-%m-%d")
salaries.sort_values(["date"], inplace=True)
#salaries[(salaries["category"] == "Спорт и досуг") & (salaries["date"].dt.year == 2024)]
categories = salaries["category"].unique()
counter = 1
print("Сферы: ")
for i in categories:
    print(i, end=', ')
    counter += 1
    if counter == 5:
        print()
        counter = 1
print('\n\n')
inp = input("Введите желаемую сферу: ")
exit = False
while not inp in categories and len(inp) != 0:      
    print(f"Не найдено сферы '{inp}'")
    inp = input("Попробуйте ввести по-другому (Enter: остановить программу): ")
if not len(inp) == 0:
    print(f"Средняя медианная зп в сфере {inp}")
    for i in range(2020, 2026):
        df = salaries[(salaries["category"] == inp) & (salaries['date'].dt.year == i)]["value"]
        print(f"В {i} году: {round(sum(df)/len(df))}")
```

### Merging

```python
#display(nocs.head())


bios_new = pd.merge(left=bios, right=nocs, left_on="born_country", right_on="NOC", how='left')
bios_new.rename(columns={"region":"born_country_full"},inplace=True)
bios_new.head()

#print(bios_new.loc[4, :])
```

```python
bios_new[(bios_new["NOC_x"] != bios_new["born_country_full"]) & (~bios_new["born_country_full"].isna())][["name" ,"NOC_x", "born_country_full"]]
```

### Concatinating DFs

```python
usa = bios[bios['born_country'] == "USA"].copy()
rus = bios[bios['born_country'] == "RUS"].copy()
#rus.head()
new_df = pd.concat([usa, rus])
new_df
```
