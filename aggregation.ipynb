# Aggregating Data

```python
import pandas as pd
bios = pd.read_csv("./data/bios.csv")
efoc = pd.read_csv("./warmup-data/coffee.csv")
efoc['Price'] = efoc['Units Sold'] * 3
```

#### value_counts

```python
bios[bios['born_country'] == 'USA']['born_region'].value_counts().head(4)
```

##### groupby

```python
efoc.groupby(['Coffee Type'])['Units Sold'].mean()
```

```python
efoc.groupby(['Coffee Type', 'Day']).agg({"Units Sold":"sum", "Price":"mean"})
```

## Pivoting

```python
pivot = efoc.pivot(columns='Coffee Type', index = "Day", values = "Price")
display(pivot)
efoc
```

## Grouping with DT (& _really_ important *RESET VALUES*)

```python
bios['born_date'] = pd.to_datetime(bios['born_date'])
bios['month_born'] = bios['born_date'].dt.month
bios.groupby([bios['born_date'].dt.year ,bios['month_born']])['name'].count().reset_index().sort_values(['name'], ascending=[0])
```
