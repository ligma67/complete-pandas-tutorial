# Advanced functionality

```python
import pandas as pd
import numpy as np
efoc = pd.read_csv("./warmup-data/coffee.csv")
bios = pd.read_csv("./data/bios.csv")
efoc['Price'] = np.where(efoc['Coffee Type'] == "Espresso", 3.99, 5.99)
efoc["Revenue"] = efoc['Price'] * efoc['Units Sold']
```

## .shift()   _(I FUCKING LOVE THIS)_

```python
efoc['yesterday_revenue'] = efoc['Revenue'].shift(2)
efoc['procent_change (%)'] = round(efoc["Revenue"]/efoc['yesterday_revenue']*100-100, 2)

efoc
```

## .rank()    (really great for ranking)

```python
bios['height_rank'] = bios['height_cm'].rank(ascending=0)
bios.sort_values(['height_rank'], ascending=[1], inplace=True)
bios
```

## .cumsum() (yes, cumsum, really good tho)

```python
efoc['cum_revenue'] = efoc['Revenue'].cumsum()
efoc
```

## .rolling()  ( kinda like cumsum() )

```python
latte = efoc[efoc["Coffee Type"] == "Latte"].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum()
latte
```
